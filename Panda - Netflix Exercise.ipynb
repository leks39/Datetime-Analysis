#### Question 1. What year has highest number of releases?

```python
#checking for number of distinct value in each column
netflix_pd.nunique()
```

```python
#checking for value in type column
netflix_pd['type'].unique()
```

```python
#sort df for only movies
movies_df = netflix_pd[netflix_pd['type'] == 'Movie']
movies_df.head()
```

```python
#['count'] = 1 allows for values to be counted
movies_df['count'] = 1
no_of_yearly_release = movies_df.groupby(['release_year']).sum().reset_index()[['release_year', 'count']]
no_of_yearly_release
```

```python
no_of_yearly_release.sort_values(['count'], ascending = False)
```

#### Question. 2 What year did Netflix add most content

```python
#converting date added column to datetime format
netflix_pd['new_date'] = pd.to_datetime(netflix_pd['date_added'])
netflix_pd.dtypes
```

```python
#applying datetime function to get year added
netflix_pd['year_added'] = netflix_pd['new_date'].dt.year
netflix_pd.head()
```

```python
netflix_pd['count'] = 1
year_content_added = netflix_pd.groupby(['year_added']).sum().reset_index()[['year_added', 'count']]
year_content_added
```

```python
year_content_added.sort_values(['count'], ascending = False)
```

#### Question 3. Most Popular Month to add content

```python
#applying datetime function to get month added
netflix_pd['month_added'] = netflix_pd['new_date'].dt.month
netflix_pd.head()
```

```python
netflix_pd['count'] = 1
month_content_added = netflix_pd.groupby(['month_added']).sum().reset_index()[['month_added', 'count']]
month_content_added
```

#### Question.5 Which actor/actress shows up most in the netflix dataframe

```python
#increase with of df for better view
pd.options.display.max_colwidth = 200

#create new df consisting of only cast
cast_df = netflix_pd[['cast']]
cast_df.head()
```

```python
#drop null values
cast_df = cast_df.dropna()
cast_df.head()
```

```python
#ensuring call cast names are in same case
cast_df['cast'] = cast_df['cast'].str.title()
cast_df.head()
```

```python
#counting names in the cast column using str.split() function and spliting on commas (,)
name_count = cast_df['cast'].str.split(',').explode().value_counts()
name_count.head(20)
```
