### Converting column to datetime format

**parse_date** argument converts specified column in this case "Date" on position 0 to a  datetime format while loading data ab initio. Another option is to used the pd.to_datetime() method. This works by parsing pd.to_datetime(column_name, format = %Y-%m-%d) or any other desired date format.

```python
avo_csv = pd.read_csv("C:\\Users\\ofagb\\OneDrive\\Documents\\Python\\avocado_full.csv", parse_dates =[0])
avo_csv
```

```python
avo_csv.dtypes
```

```python
### adding day_name, month name, week, quarter to df
avo_csv['Day_name'] = avo_csv['Date'].dt.day_name()
avo_csv['Month_name'] = avo_csv['Date'].dt.month_name()
avo_csv['Week'] = avo_csv['Date'].dt.week
avo_csv['Quarter'] = avo_csv['Date'].dt.quarter
avo_csv['month_year'] = avo_csv['Date'].dt.to_period('M')
```

```python
#view new df
avo_csv.head()
```

```python
## Extracting info from date column then converting to a new df
date_column = avo_csv['Date']

dates = pd.DataFrame({"year": date_column.dt.year,"month": date_column.dt.month,"month_name": date_column.dt.month_name(),
                      "day": date_column.dt.day,"day_name": date_column.dt.day_name(), "dayofyear": date_column.dt.dayofyear,
                      "week": date_column.dt.week,  "weekofyear": date_column.dt.weekofyear,  "dayofweek": date_column.dt.dayofweek,
                      "quarter": date_column.dt.quarter})
dates
```

```python
#earliest date
date_column.min()
```

```python
#Time Delta i.e time between two dates can be gotten by subsctraction in Python
time_diff = date_column.max() - date_column.min()
time_diff
```
